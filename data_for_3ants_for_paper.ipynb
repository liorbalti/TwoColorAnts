```python
%matplotlib qt
data_dict={'time_s' : exp11.bdata[['time']]}
for ant_id in [289,148,418]:
    if ant_id in exp11.foragers_list:
        ant = Data.ForagerData(ant_id, feedings_df=exp11.feedings_df, bdata_df=exp11.bdata, interactions_df=exp11.interactions_df)
    else:
        ant = Data.AntData(ant_id, bdata_df=exp11.bdata, interactions_df=exp11.interactions_df)
    
    raw_crop_df = pd.DataFrame.from_dict(ant.crop_dict_raw)
    raw_crop_df = raw_crop_df.rename(columns={'red':f'a{ant_id}- raw crop red', 'yellow':f'a{ant_id}- raw crop yellow'})
    
    clean_crop_df = ant.clean_crop()
    clean_crop_df = clean_crop_df.rename(columns={'red':f'a{ant_id}- clean crop red (intensity)','yellow':f'a{ant_id}- clean crop yellow (intensity)'})
    for color in ['red','yellow']:
        k = exp11.conversion_factors_by_weights_df.loc[0,color]
        clean_crop_df[f'a{ant_id}- clean crop {color} (ul)'] = clean_crop_df[f'a{ant_id}- clean crop {color} (intensity)']/k
    
    ant_data = [ant.x_raw,ant.raw_y,raw_crop_df,clean_crop_df]
    data_dict.update({ant_id: pd.concat(ant_data,axis=1)})

data_df = pd.concat(data_dict,axis=1)
#data_df.columns=[f'a{ant}-{col}' for ant, col in data_df.columns.values]
data_df.head()
```
