```python
import pandas as pd
import numpy as np
import Data
import matplotlib.pyplot as plt
from os import sep as sep
```

```python
exp11 = Data.ExperimentData(11, bdata_path='blob analysis normalized by white paper')
```

```python
%matplotlib qt
clean_crop_dict={}
for ant_id in exp11.ants_list:
    if ant_id in exp11.foragers_list:
        ant = Data.ForagerData(ant_id, feedings_df=exp11.feedings_df, bdata_df=exp11.bdata, interactions_df=exp11.interactions_df)
    else:
        ant = Data.AntData(ant_id, bdata_df=exp11.bdata, interactions_df=exp11.interactions_df)
    
    clean_crop_ant = ant.clean_crop()
    # ant.plot_raw_timeline()
    # plt.plot(clean_crop_ant)
    # plt.show()
    
    clean_crop_dict.update({ant_id: clean_crop_ant})

clean_crop_df = pd.concat(clean_crop_dict,axis=1)
clean_crop_df.columns=[f'a{ant}-{col}' for ant, col in clean_crop_df.columns.values]
clean_crop_df.head()
```
