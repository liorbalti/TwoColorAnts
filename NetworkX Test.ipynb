```python
#Model Selection
#model = ep.SIRModel(DGWcolA)
#ERGcolA = nx.erdos_renyi_graph(76, 0.088245614) #number of nodes, percentage of possible edges realized
model = ep.SIRModel(DGWcolA)
#nx.draw(ERGcolA)
```

```python
#Model Configuration for Network Diffusion
config = mc.Configuration()
infected_nodes = [171, 180, 207, 268, 421] #These are the foragers in Colony A.
#infected_nodes = [148, 424, 454] #These are the foragers in Colony B.
#infected_nodes = [302, 365, 95, 291] #These are the foragers in Colony C.
#infected_nodes = [1, 2, 3, 4, 5] #These are the foragers in the Erdos-Renyi Graph Analog of Colony A.
#infected_nodes = [1, 2, 3] #These are the foragers in the Erdos-Renyi Graph Analog of Colony B.
#infected_nodes = [1, 2, 3, 4] #These are the foragers in the Erdos-Renyi Graph Analog of Colony C.
config.add_model_initial_configuration("Infected", infected_nodes)
config.add_model_parameter('beta', 0.001) #Infection Probability 0.001 (for each ant interaction, P of trophallaxis)
config.add_model_parameter('gamma', 0) #Removal Probability 0.01 (for trophallaxis, P of never having future interactions)
config.add_model_parameter("fraction_infected", 0.013) # 0.05 (% receiving trophallaxis with each interaction: 1/(N-1))
model.set_initial_status(config)
```

```python
#Simulation
iterations = model.iteration_bunch(2000)
trends = model.build_trends(iterations)
```

```python
#Visualization Diffusion Trend
from bokeh.io import output_notebook, show
from ndlib.viz.bokeh.DiffusionTrend import DiffusionTrend
viz = DiffusionTrend(model, trends)
p=viz.plot(width=400, height=400)
show(p)
```
